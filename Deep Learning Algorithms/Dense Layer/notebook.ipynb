# Dense Layer from Scratch

```python
import cupy as cp
import nnfs
from nnfs.datasets import spiral_data
import matplotlib.pyplot as plt
nnfs.init()
```

```python
x, y = spiral_data(samples=100, classes=3)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, cmap="YlOrBr")
plt.title("spiral dataset");
```

```python
# Dense Layer
class DenseLayer:

    # Layer Initialization
    def __init__(self, num_inputs, num_neurons):
        # Initialize weights and biases
        self.output = None
        self.weights = 0.01 * cp.random.randn(num_inputs, num_neurons)
        self.biases = cp.zeros((1, num_neurons))

    # Forward Pass
    def forward(self, inputs):
        # Calculate output values from inputs, weights and biases
        self.output = cp.dot(inputs, self.weights) + self.biases
```

```python
dense_layer = DenseLayer(2, 3) # 2 input features and 3 output values
```

```python
x = cp.array(x)
y = cp.array(y)
```

```python
dense_layer.forward(x)
```
