## Measuring Models and Performance

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset
import scipy.stats as stats
import torch.nn.functional as F
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({"font.size": 18})
```

### Runnign experiments

```python
n = 50
points = 10000
acc = np.zeros(points)
pre = np.zeros(points)
rec = np.zeros(points)
f1 = np.zeros(points)

for i in range(points):
    tp = np.random.randint(1, n)
    fn = n - tp
    tn = np.random.randint(1, n)
    fp = n - tn
    acc[i] = (tp + tn) / (2 * n)
    pre[i] = tp / (tp + fp)
    rec[i] = tp/(tp + fn)
    f1[i] = tp / (tp + (fp + fn) / 2)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

ax[0].scatter(acc, f1, s=5, c=pre)
ax[0].plot([0, 1], [.5, .5], "k--", linewidth=.5)
ax[0].plot([.5, .5], [0, 1], "k--", linewidth=.5)
ax[0].set_xlabel("accuracy")
ax[0].set_ylabel("f1")
ax[0].set_title("f1 by precision")

ax[1].scatter(acc, f1, s=5, c=rec)
ax[1].plot([0, 1], [.5, .5], "k--", linewidth=.5)
ax[1].plot([.5, .5], [0, 1], "k--", linewidth=.5)
ax[1].set_xlabel("accuracy")
ax[1].set_ylabel("f1")
ax[1].set_title("f1 by rec")
plt.show()
```

### Doing the same with wine quality dataset

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv",
                  sep=';')
```

```python
data.head()
```

```python
data = data[data["total sulfur dioxide"] < 200]
zscore = data.keys()
zscore = zscore.drop("quality")
data[zscore] = data[zscore].apply(stats.zscore)
data["good"] = 0
data["good"][data["quality"] > 5] = 1
```

```python
data.head()
```

```python
data_t = torch.tensor(data[zscore].values).float()
labels = torch.tensor(data["good"].values).float()
train_data, test_data, train_labels, test_labels = train_test_split(data_t,
                                                                   labels,
                                                                   test_size=0.1)
```

```python
train_data = TensorDataset(train_data, train_labels)
test_data = TensorDataset(test_data, test_labels)

train_loader = DataLoader(train_data,
                         batch_size=32,
                         shuffle=True,
                         drop_last=True)
test_loader = DataLoader(test_data,
                        batch_size=test_data.tensors[0].shape[0])
```

### Creating deep learning model

```python
class WineNet(nn.Module):
    def __init__(self):
        super().__init__()
        # input layer
        self.input = nn.Linear(12, 16)
        
        # hidden layers
        self.hidden_1 = nn.Linear(16, 32)
        self.hidden_2 = nn.Linear(32, 32)
        # output layer
        self.output = nn.Linear(32, 1)
    
    # forward pass
    def forward(self, x):
        x = F.relu(self.input(x))
        x = F.relu(self.hidden_1(x))
        x = F.relu(self.hidden_2(x))
        return self.output(x)
```

```python
def compute_acc(yh, y):
    true = (((yh > 0) == y).float()).item()
    return torch.mean(true)
```

```python
def train_model(epochs=1000):
    train_acc = []
    test_acc = []
    train_loss = []
    model = WineNet()
    Loss = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
    
    for i in range(epochs):
        batch_acc = []
        batch_loss = []
        for x, y in train_loader:
            # forward pass
            yh = model(x)
            loss = Loss(yh, y)
            # computing backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            # computing performance
            batch_loss.append(loss.item())
            batch_acc.append(compute_acc(yh, y))
        # computing overall performance
        train_acc.append(np.mean(batch_acc))
        train_loss.append(np.mean(batch_loss))
        # computing test performance
        x, y = next(iter(test_loader))
        with torch.no_grad():
            yh = model(x)
        test_acc(yh, y)
    return train_acc, test_acc, train_loss, model
```

### Creating model and fitting model

```python
train_acc, test_acc, train_loss, model = train_model()
```
